```python
size = 512
batchsize = 8
tfms = get_transforms(do_flip = True)
src = (ImageImageList.from_folder(image_path)
       .split_by_rand_pct(0.1,seed=2).label_from_func(lambda x: x))
data = (src.transform(tfms, size=size, tfm_y=True)
        .databunch(bs=batchsize)
        .normalize(imagenet_stats, do_y = False))
```

```python
data
```

```python
data.show_batch(2, figsize=(10,7))
```

```python
!mkdir ./models
```

```python
acc = partial(accuracy_thresh, thresh=0.1)

learn = unet_learner(data, models.resnet34, loss_func=F.mse_loss, metrics=acc)
learn.model_dir = Path('/kaggle/working/models')
```

```python
learn.load('autoenc{}'.format(i))
i += 1
```

```python
learn.summary()
```

```python
learn.unfreeze()
learn.lr_find()
learn.recorder.plot()
```

```python
learn.unfreeze()
learn.fit_one_cycle(5, slice(1e-6))
```

```python
learn.save('autoenc{}'.format(i))
```
